```python
# Install dependencies if needed
# !pip install pgmpy lingam

from pgmpy.utils import get_example_model
import pandas as pd
import numpy as np

# Load a small built-in Bayesian network
# 'sachs' is a well-known 11-node protein signalling network
true_model = get_example_model("sachs")

# Simulate continuous data
# LiNGAM requires continuous (not discrete) data
sachs_df = true_model.simulate(n_samples=1000, show_progress=False)

print("Loaded model with", len(true_model.nodes()), "nodes and", len(true_model.edges()), "edges")
sachs_df.head()
```

```python
# pgmpy simulates discrete data by default.
# We add small Laplace noise to make it continuous and non-Gaussian,
# which is the key assumption LiNGAM relies on.

rng = np.random.default_rng(42)
noise = rng.laplace(loc=0, scale=0.3, size=sachs_df.shape)
sachs_continuous = sachs_df.astype(float) + noise

print(sachs_continuous.shape)
sachs_continuous.describe().round(3)
```
